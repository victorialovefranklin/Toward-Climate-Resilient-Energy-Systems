```python
# ==============================
# California Grid: Comparative Analysis Framework
# Author: You + ChatGPT
# ==============================

import os, re, math, warnings
from pathlib import Path
import numpy as np
import pandas as pd

warnings.filterwarnings("ignore")

# -------------------------
# 0) CONFIG & HELPERS
# -------------------------
DATA = {
    "eaglei_events": "/content/Eaglei_CA_Outages_with_events.csv",
    "eaglei_merged": "/content/Eaglei_CA_Outages_Merged.csv",
    "eaglei_agg":    "/content/Eaglei_CA_Outages_Agg.csv",
    "eia861":        "/content/EIA861_CA_Combined_Data.csv",
    "psps":          "/content/CA_PSPS_Cleaned_Data.csv",
    "current":       "/content/CA_Current_Power_Outages.csv",
}
OUTDIR = Path("/content/outputs")
OUTDIR.mkdir(parents=True, exist_ok=True)

def _snake(s: str) -> str:
    s = re.sub(r"[^\w]+", "_", s.strip())
    s = re.sub(r"_+", "_", s)
    return s.strip("_").lower()

def load_csv(path, dtype=None, parse_dates=None):
    try:
        df = pd.read_csv(path, dtype=dtype or {}, parse_dates=parse_dates or [])
        df.columns = [_snake(c) for c in df.columns]
        print(f"Loaded {Path(path).name}: {df.shape}")
        return df
    except Exception as e:
        print(f"Could not load {path}: {e}")
        return pd.DataFrame()

def saveout(df: pd.DataFrame, name: str):
    if df is None or df.empty:
        print(f"[skip] {name} -> empty")
        return None
    out = OUTDIR / name
    df.to_csv(out, index=False)
    print(f"[saved] {name}  ->  {out}")
    try:
        # Print head without triggering potential display issues in some environments
        print(df.head(2).to_markdown(index=False))
    except Exception:
        pass
    return out

def ensure_fips(series):
    # Robust FIPS standardization to 5-char string
    def _pad(x):
        if pd.isna(x): return np.nan
        s = str(x).split(".")[0]
        s = re.sub(r"[^\d]", "", s)
        if s == "": return np.nan
        return s.zfill(5)
    return series.apply(_pad)

def coalesce(df, new_col, candidates, default=np.nan):
    for c in candidates:
        if c in df.columns:
            df[new_col] = df[new_col].where(df[new_col].notna(), df[c])
    if new_col not in df.columns:
        df[new_col] = default
    return df

def safe_div(a, b):
    return np.where((b==0) | pd.isna(b), np.nan, a / b)

# -------------------------
# 1) LOAD & STANDARDIZE
# -------------------------
eaglei_events = load_csv(DATA["eaglei_events"])
eaglei_merged = load_csv(DATA["eaglei_merged"])
eaglei_agg    = load_csv(DATA["eaglei_agg"])
eia861        = load_csv(DATA["eia861"])
psps          = load_csv(DATA["psps"])
current       = load_csv(DATA["current"])


# --- Standardize key columns where possible ---
# Dates
date_cols_guess = ["start_time","end_time","run_start_time","run_end_time","start","end","event_start","event_end","timestamp","date","datetime"]
for df in [eaglei_events, psps, current, eaglei_merged, eaglei_agg]:
    if df is not None and not df.empty:
        for c in df.columns:
            if any(k in c for k in ["time","date","start","end","ts"]):
                # try parsing - leave as-is if already datetime
                if not np.issubdtype(df[c].dtype, np.datetime64):
                    with warnings.catch_warnings():
                        warnings.simplefilter("ignore")
                        try:
                            df[c] = pd.to_datetime(df[c], errors="coerce")
                        except Exception:
                            pass


# Year / month / duration
if eaglei_events is not None and not eaglei_events.empty:
    if "run_start_time" in eaglei_events.columns:
        eaglei_events["year"] = eaglei_events["run_start_time"].dt.year
        eaglei_events["month"] = eaglei_events["run_start_time"].dt.month
    elif "start_time" in eaglei_events.columns:
        eaglei_events["year"] = eaglei_events["start_time"].dt.year
        eaglei_events["month"] = eaglei_events["start_time"].dt.month

# FIPS / county
for df in [eaglei_events, eaglei_merged, eaglei_agg, psps, current, eia861]:
    if df is not None and not df.empty:
        if "fips" in df.columns:
            df["county_fips"] = ensure_fips(df["fips"])
        elif "county_fips" in df.columns:
            df["county_fips"] = ensure_fips(df["county_fips"])
        # common county name fields
        if "county" in df.columns:
            df["county_name"] = df["county"].astype(str).str.strip()
        elif "county_name" in df.columns:
            df["county_name"] = df["county_name"].astype(str).str.strip()

# Customers affected / totals
# EAGLE-I
cand_out = ["customers_out","customers_affected","customersaffected","cust_out","cust_affected"]
if eaglei_events is not None and not eaglei_events.empty:
    eager_out = None
    for c in cand_out:
        if c in eaglei_events.columns and eaglei_events[c].notna().any():
            eager_out = c
            break
    if eager_out is not None:
        eaglei_events.rename(columns={eager_out: "customers_affected"}, inplace=True)
    elif "mean_customers" in eaglei_events.columns:
         eaglei_events["customers_affected"] = pd.to_numeric(eaglei_events["mean_customers"], errors="coerce")
    else:
        eaglei_events["customers_affected"] = np.nan


# EIA sector totals (counts)
# Try to map multiple possible names
EIA_MAP = {
    "residential_customers": ["residential_customer_count"],
    "commercial_customers":  ["commercial_customer_count"],
    "industrial_customers":  ["industrial_customer_count"],
    "transport_customers":   ["transportation_customer_count"],
    "total_customers":       ["total_customer__count"], # Corrected mapping for the space
    "utility_id":            ["utility_id","utility_number","utilityid","respondent_id"],
    "year":                  ["year","data_year","report_year"]
}
if eia861 is not None and not eia861.empty:
    for new_col, opts in EIA_MAP.items():
        if new_col not in eia861.columns:
            for cand in opts:
                if cand in eia861.columns:
                    eia861.rename(columns={cand: new_col}, inplace=True)
                    break

    # Ensure numeric, handling '.' as NaN
    for c in ["residential_customers","commercial_customers","industrial_customers","transport_customers","total_customers"]:
        if c in eia861.columns:
            eia861[c] = eia861[c].astype(str).str.replace(",", "", regex=False).replace(".", np.nan).astype(float)


# -------------------------
# 2) DERIVED FIELDS
# -------------------------
# Duration in minutes for EAGLE-I events (if start/end available)
def add_duration_minutes(df):
    if df is None or df.empty:
        return df
    candidates = [
        ("run_start_time","run_end_time"),
        ("start_time","end_time"),
        ("event_start","event_end"),
        ("start","end"),
    ]
    for st, en in candidates:
        if st in df.columns and en in df.columns:
            # Ensure datetime objects before calculating difference
            try:
                df[st] = pd.to_datetime(df[st], errors='coerce')
                df[en] = pd.to_datetime(df[en], errors='coerce')
                dur = (df[en] - df[st]).dt.total_seconds() / 60.0
                df["duration_min"] = dur
                return df
            except Exception:
                pass # Continue to next candidate if parsing fails
    # fallback: none
    if "duration_min" not in df.columns:
        df["duration_min"] = np.nan
    return df


eaglei_events = add_duration_minutes(eaglei_events)


# Sector shares (EIA)
if eia861 is not None and not eia861.empty:
    if "total_customers" in eia861.columns and eia861["total_customers"].notna().any():
        for c in ["residential_customers","commercial_customers","industrial_customers","transport_customers"]:
            if c in eia861.columns:
                eia861[c+"_share"] = safe_div(eia861[c], eia861["total_customers"])

# -------------------------
# 3) ANALYSIS FUNCTIONS
# -------------------------

# === I. Temporal & Trend ===
def annual_outage_trends(df):
    if df is None or df.empty or "year" not in df.columns:
        return pd.DataFrame()
    out_df = (df
          .groupby("year", dropna=True)
          .agg(outage_events=("year","count"),
               customers_affected=("customers_affected","sum"),
               avg_duration_min=("duration_min","mean"))
          .reset_index()
         )
    saveout(out_df, "annual_outage_trends.csv")
    return out_df

def seasonal_patterns(df):
    if df is None or df.empty or not {"year","month"}.issubset(df.columns):
        return None
    out_df = (df
          .groupby(["year","month"], dropna=True)
          .agg(events=("month","count"),
               customers_affected=("customers_affected","sum"),
               avg_duration_min=("duration_min","mean"))
          .reset_index()
         )
    saveout(out_df, "seasonal_patterns.csv")
    return out_df

def duration_trends(df):
    if df is None or df.empty or "year" not in df.columns:
        return None
    out_df = (df
          .groupby("year", dropna=True)
          .agg(avg_duration_min=("duration_min","mean"),
               p50_duration_min=("duration_min","median"),
               p90_duration_min=("duration_min", lambda s: s.quantile(0.90)))
          .reset_index()
         )
    return saveout(out_df, "duration_trends.csv")

def yoy_growth(df):
    if df is None or df.empty: return None
    # Use the DataFrame directly
    cols = ["outage_events","customers_affected"]
    for c in cols:
        if c in df.columns:
            df[f"{c}_yoy_pct"] = df[c].pct_change() * 100.0
    return saveout(df, "yoy_growth_metrics.csv")


# === II. Coverage & Representation ===
def utility_coverage_completeness(eia_df, merged_df):
    # Requires merged dataset connecting outages->utilities, else approximate by count of utilities per year in eia861
    if eia_df is None or eia_df.empty:
        return None
    df = eia_df.copy()
    util_per_year = df.groupby("year")["utility_id"].nunique().reset_index(name="utilities_eia861")
    # if eaglei_merged has utility_id, compare
    if merged_df is not None and not merged_df.empty and "utility_id" in merged_df.columns and "year" in merged_df.columns:
        u2 = merged_df.groupby("year")["utility_id"].nunique().reset_index(name="utilities_eaglei")
        out = util_per_year.merge(u2, on="year", how="outer")
        out["coverage_ratio"] = safe_div(out["utilities_eaglei"], out["utilities_eia861"])
    else:
        out = util_per_year
    return saveout(out, "utility_coverage_completeness.csv")

def county_representation(event_df, psps_df, eia_df):
    # Compare counties represented in each source by year (where available)
    def by_year_counties(df, src_name):
        if df is not None and not df.empty and "year" in df.columns and ("county_fips" in df.columns or "county_name" in df.columns):
            key = "county_fips" if "county_fips" in df.columns else "county_name"
            out = (df.groupby("year")[key]
                   .nunique()
                   .reset_index(name=f"{src_name}_counties"))
            return out
        return pd.DataFrame()

    parts = []
    parts.append(by_year_counties(event_df, "eaglei"))
    parts.append(by_year_counties(psps_df, "psps"))
    parts.append(by_year_counties(eia_df, "eia861"))  # service presence by year
    parts = [p for p in parts if not p.empty]
    if not parts: return None
    out = parts[0]
    for p in parts[1:]:
        out = out.merge(p, on="year", how="outer")
    return saveout(out, "county_representation.csv")

# === III. Sectoral & Customer-Type ===
def sectoral_distribution(eia_df):
    if eia_df is None or eia_df.empty:
        return pd.DataFrame() # Return empty DataFrame if input is None or empty
    cols = ["residential_customers","commercial_customers","industrial_customers","transport_customers","total_customers"]
    # Check if all required columns exist before proceeding
    if not all(c in eia_df.columns for c in cols):
        print(f"Missing required columns for sectoral_distribution: {list(set(cols) - set(eia_df.columns))}")
        return pd.DataFrame()
    views = (eia_df.groupby("year")[cols].sum(min_count=1).reset_index())
    for c in cols[:-1]:
        if c in views.columns and "total_customers" in views.columns:
            views[c+"_pct"] = safe_div(views[c], views["total_customers"]) * 100.0
    return saveout(views, "sectoral_distribution_by_year.csv")

def sectoral_intensity_rate(event_df, eia_df):
    # Normalize EAGLE-I affected customers by EIA total customers (per year)
    if event_df is None or event_df.empty or "year" not in event_df.columns or eia_df is None or eia_df.empty or "total_customers" not in eia_df.columns:
        return None
    total_by_year = eia_df.groupby("year")["total_customers"].sum(min_count=1).reset_index()
    aff_by_year = event_df.groupby("year")["customers_affected"].sum(min_count=1).reset_index()
    df = aff_by_year.merge(total_by_year, on="year", how="left")
    if "customers_affected" in df.columns and "total_customers" in df.columns:
        df["outage_rate_per_1000"] = safe_div(df["customers_affected"], df["total_customers"]) * 1000.0
    return saveout(df, "sectoral_intensity_rate.csv")

def peak_sectoral_month(df):
    # If sectoral allocation exists in eaglei_merged (e.g., *_out columns), otherwise placeholder via total
    if df is None or df.empty or "month" not in df.columns:
        return None
    if "customers_affected" not in df.columns:
        return None

    agg = (df
           .groupby(["year","month"])
           ["customers_affected"].sum(min_count=1) # Select the column before sum
           .reset_index(name="customers_affected"))

    if agg.empty: return None

    # Ensure 'year' column exists in agg for idxmax
    if 'year' not in agg.columns:
         return None

    # Find the month with max customers_affected for each year
    idx = agg.groupby("year")["customers_affected"].idxmax()
    # Filter out NaN indices before using .loc
    valid_idx = idx.dropna()
    if valid_idx.empty:
        print("No valid customers_affected data to determine peak month.")
        return None
    peaks = agg.loc[valid_idx].sort_values("year")

    return saveout(peaks, "peak_outage_month_by_year.csv")


# === IV. Spatial ===
def county_hotspots(df):
    # Rank counties by cumulative customers_affected & events
    if df is None or df.empty or "customers_affected" not in df.columns: return pd.DataFrame()
    key = "county_fips" if "county_fips" in df.columns else "county_name"
    if key not in df.columns: return pd.DataFrame()
    df = (df.groupby(key)
          .agg(events=("duration_min","count"),
               customers_affected=("customers_affected","sum"),
               avg_duration_min=("duration_min","mean"))
          .reset_index()
          .sort_values(["customers_affected","events"], ascending=[False, False])
         )
    return saveout(df, "county_hotspots.csv")

def regional_compare(df):
    # Simple coastal vs inland using county_name heuristic (you can replace with real lookup)
    if df is None or df.empty or "county_name" not in df.columns:
        return None
    coastal = set([
        # quick coastal list (customize as needed)
        "los angeles","orange","san diego","ventura","santa barbara","san luis obispo",
        "monterey","santa cruz","san mateo","san francisco","marin","sonoma","mendocino",
        "humboldt","del norte"
    ])
    tmp = df.copy()
    tmp["region"] = np.where(tmp["county_name"].str.lower().isin(coastal), "coastal","inland")
    df = (tmp.groupby("region")
          .agg(events=("duration_min","count"),
               customers_affected=("customers_affected","sum"),
               avg_duration_min=("duration_min","mean"))
          .reset_index())
    return saveout(df, "regional_coastal_vs_inland.csv")

# === V. Event-Type & Cause ===
def event_type_distribution(event_df, psps_df):
    # DOE-417 types living in eaglei_events? fallback to psps cause
    src = None
    if event_df is not None and not event_df.empty and "event_type" in event_df.columns:
        src = (event_df.groupby("event_type")
               .agg(events=("event_type","count"),
                    customers_affected=("customers_affected","sum"),
                    avg_duration_min=("duration_min","mean"))
               .reset_index()
              )
        return saveout(src, "doe417_event_type_distribution.csv")
    elif psps_df is not None and not psps_df.empty and "cause" in psps_df.columns:
        src = psps_df.groupby("cause").size().reset_index(name="events")
        return saveout(src, "psps_cause_distribution.csv")
    return None


def psps_alignment(event_df, psps_df):
    # Align PSPS dates with EAGLE-I spikes (monthly)
    if event_df is None or event_df.empty or psps_df is None or psps_df.empty:
        return None

    event_ts_col = None
    for c in ["run_start_time", "start_time"]:
        if c in event_df.columns:
            event_ts_col = c
            break
    if event_ts_col is None: return None

    psps_ts_col = None
    for c in ["date", "datetime", "start", "start_time"]:
        if c in psps_df.columns:
            psps_ts_col = c
            break
    if psps_ts_col is None: return None

    e_df = event_df.copy()
    e_df["ym"] = pd.to_datetime(e_df[event_ts_col], errors='coerce').dt.to_period("M")

    p_df = psps_df.copy()
    p_df["ym"] = pd.to_datetime(p_df[psps_ts_col], errors='coerce').dt.to_period("M")


    m1 = (e_df.groupby("ym")["customers_affected"].sum(min_count=1)
          .reset_index(name="eaglei_customers"))
    m2 = p_df.groupby("ym").size().reset_index(name="psps_events")
    out = m1.merge(m2, on="ym", how="outer").sort_values("ym")
    return saveout(out, "psps_alignment_monthly.csv")


# === VI. Ratios & Stats ===
def outage_rate_per_1000(event_df, eia_df):
    return sectoral_intensity_rate(event_df, eia_df)

def peak_to_average_ratio(df):
    # Monthly PAR across whole state
    if df is None or df.empty:
        return None

    tcol = None
    for c in ["run_start_time", "start_time"]:
        if c in df.columns:
            tcol = c
            break
    if tcol is None:
        return None

    df["ym"] = pd.to_datetime(df[tcol], errors='coerce').dt.to_period("M")
    m = df.groupby("ym")["customers_affected"].sum(min_count=1)

    if m.empty or m.mean() == 0:
        return None

    par = pd.DataFrame({
        "peak": [m.max()],
        "average": [m.mean()],
        "peak_to_average_ratio": [m.max()/m.mean()]
    })
    return saveout(par, "peak_to_average_ratio.csv")


def correlations_monthly(event_df, psps_df):
    # Monthly correlations: EAGLE-I customers vs PSPS events
    pa = psps_alignment(event_df, psps_df)
    if pa is None: return None
    # pa is the result of saveout, which saves to a file.
    # We need to load the saved file to get the dataframe.
    try:
        df = pd.read_csv(OUTDIR / "psps_alignment_monthly.csv")
    except Exception:
        return None

    if df.empty or "eaglei_customers" not in df.columns or "psps_events" not in df.columns:
        return None

    corr = df[["eaglei_customers","psps_events"]].corr(method="pearson")
    out = corr.reset_index().rename(columns={"index":"metric"})
    return saveout(out, "correlations_monthly_eaglei_psps.csv")

# === VII. Reliability Proxies (SAIDI/SAIFI-like) ===
def reliability_proxies(event_df, eia_df):
    # SAIFI ~ outages per customer; SAIDI ~ minutes of interruption per customer
    if event_df is None or event_df.empty or "year" not in event_df.columns or eia_df is None or eia_df.empty:
        return pd.DataFrame() # Return empty DataFrame if input is None or empty

    # Check if total_customers column exists before grouping
    if "total_customers" not in eia_df.columns:
        print("Missing required column 'total_customers' for reliability_proxies.")
        return pd.DataFrame()

    tot_customers_year = eia_df.groupby("year")["total_customers"].sum(min_count=1).reset_index()
    by_year = (event_df
               .groupby("year")
               .agg(total_events=("duration_min","count"),
                    total_minutes=("duration_min","sum"),
                    total_customers_affected=("customers_affected","sum"))
               .reset_index())

    df = by_year.merge(tot_customers_year, on="year", how="left")

    if "total_events" in df.columns and "total_customers" in df.columns:
        df["saifi_proxy"] = safe_div(df["total_events"], df["total_customers"])  # events per customer
    if "total_minutes" in df.columns and "total_customers" in df.columns:
        df["saidi_proxy_min"] = safe_div(df["total_minutes"], df["total_customers"])  # minutes per customer

    return saveout(df, "reliability_proxies.csv")

# === VIII. Policy / Reporting ===
def policy_impact_prepost_2019(df):
    # Compare pre-2019 vs 2019+ (PSPS regulation era)
    if df is None or df.empty or "year" not in df.columns:
        return None
    df = (df
          .assign(period=lambda x: np.where(x["year"] < 2019, "pre_2019","2019_plus"))
          .groupby("period")
          .agg(events=("duration_min","count"),
               customers_affected=("customers_affected","sum"),
               avg_duration_min=("duration_min","mean"))
          .reset_index())
    return saveout(df, "policy_impact_prepost_2019.csv")


def utility_benchmarking(merged_df, eia_df):
    # Requires utility_id in merged file
    if merged_df is None or merged_df.empty or not {"utility_id","year"}.issubset(merged_df.columns) or eia_df is None or eia_df.empty:
        return None
    # Try to find outage intensity columns in merged
    cols = [c for c in merged_df.columns if "customers" in c and ("out" in c or "affected" in c)]
    use = cols if cols else ["customers_affected"]
    use = [c for c in use if c in merged_df.columns]
    if not use: return None
    agg = (merged_df
           .groupby(["utility_id","year"])
           [use].sum(min_count=1)
           .reset_index())
    # Normalize by EIA totals per utility-year if available
    base = eia_df.groupby(["utility_id","year"])["total_customers"].sum(min_count=1).reset_index()
    out = agg.merge(base, on=["utility_id","year"], how="left")
    for c in use:
        if c in out.columns and "total_customers" in out.columns:
            out[c+"_per_1000"] = safe_div(out[c], out["total_customers"]) * 1000.0
    return saveout(out, "utility_benchmarking.csv")

def documentation_rate(df):
    # % of EAGLE-I events with an event_type (DOE-417-like tag)
    if df is None or df.empty: return None
    has = "event_type" in df.columns
    by_year = (df
               .assign(has_event_type = has & df.get("event_type").notna() if has else False)
               .groupby("year")
               .agg(total=("duration_min","count"),
                    documented=("has_event_type","sum"))
               .reset_index())
    if "total" in by_year.columns and "documented" in by_year.columns:
        by_year["documentation_rate_pct"] = safe_div(by_year["documented"], by_year["total"]) * 100.0
    return saveout(by_year, "documentation_rate.csv")

# === IX. Forecasting / ML (placeholders to wire later) ===
def simple_baseline_projection(df):
    # Naive: last 3-year mean as projection for next year
    if df is None or df.empty: return None
    df = df.dropna(subset=["year"])
    df = df.sort_values("year")
    if len(df) < 3:
        return saveout(df.assign(note="Not enough years for baseline"), "baseline_projection.csv")
    window = df.tail(3)
    proj = {
        "metric": [],
        "projection_next_year": []
    }
    for metric in ["outage_events","customers_affected","avg_duration_min"]:
        if metric in window.columns:
             proj["metric"].append(metric)
             proj["projection_next_year"].append(window[metric].mean())

    out = pd.DataFrame(proj)

    return saveout(out, "baseline_projection.csv")


# -------------------------
# 4) RUNNER (pick & choose)
# -------------------------
if __name__ == "__main__":
    print("\n=== I. Temporal & Trend ===")
    annual_trends_df = annual_outage_trends(eaglei_events)
    seasonal_patterns(eaglei_events)
    duration_trends(eaglei_events)
    yoy_growth(annual_trends_df)

    print("\n=== II. Coverage & Representation ===")
    utility_coverage_completeness(eia861, eaglei_merged)
    county_representation(eaglei_events, psps, eia861)

    print("\n=== III. Sectoral ===")
    sectoral_distribution(eia861)
    sectoral_intensity_rate(eaglei_events, eia861)
    peak_sectoral_month(eaglei_events)

    print("\n=== IV. Spatial ===")
    county_hotspots(eaglei_events)
    regional_compare(eaglei_events)

    print("\n=== V. Event-Type & Cause ===")
    event_type_distribution(eaglei_events, psps)
    psps_alignment(eaglei_events, psps)

    print("\n=== VI. Ratios & Stats ===")
    outage_rate_per_1000(eaglei_events, eia861)
    peak_to_average_ratio(eaglei_events)
    correlations_monthly(eaglei_events, psps)

    print("\n=== VII. Reliability Proxies ===")
    reliability_proxies(eaglei_events, eia861)

    print("\n=== VIII. Policy / Reporting ===")
    policy_impact_prepost_2019(eaglei_events)
    utility_benchmarking(eaglei_merged, eia861)
    documentation_rate(eaglei_events)

    print("\n=== IX. Forecasting / ML (baseline) ===")
    simple_baseline_projection(annual_trends_df)

    print("\nAll outputs saved under:", OUTDIR)
```

Prepare the outage data by loading `county_hotspots.csv` and `eia861.csv`, merging them to calculate outages per 1,000 customers, and filtering for the latest year.

**Reasoning**:
Load the county outage data and EIA customer data, merge them on county and year, calculate the outage rate per 1,000 customers, and filter for the latest year to prepare the data for the choropleth map.

```python
import pandas as pd

try:
    # Load county hotspots data
    county_hotspots_df = pd.read_csv('/content/outputs/county_hotspots.csv')

    # Load EIA customer data
    # Assuming eia861 DataFrame is already loaded from previous steps,
    # otherwise load it here:
    # eia861 = load_csv(DATA["eia861"]) # Ensure load_csv is defined or use pd.read_csv with appropriate cleaning

    # Calculate total customers per county and year from eia861
    # Need to ensure 'county_fips' and 'total_customers' are available and clean in eia861
    if eia861 is None or eia861.empty or 'county_fips' not in eia861.columns or 'total_customers' not in eia861.columns:
         print("Error: EIA data (eia861) is not loaded or missing required columns.")
         outage_customer_data = pd.DataFrame() # Create empty df to avoid errors
    else:
        eia_county_year = eia861.groupby(['year', 'county_fips'])['total_customers'].sum(min_count=1).reset_index(name='total_customers')

        # Merge county hotspots with EIA customer data
        outage_customer_data = pd.merge(county_hotspots_df, eia_county_year, on='county_fips', how='left')

        # Calculate outages per 1000 customers
        # Use events for rate calculation
        outage_customer_data['outages_per_1000_customers'] = (outage_customer_data['events'] / outage_customer_data['total_customers']) * 1000

        # Get the latest year from the county_hotspots data
        latest_year = county_hotspots_df['year'].max()

        # Filter data for the latest year - need 'year' in county_hotspots_df
        # The county_hotspots_df does not have a year column directly, it's aggregated.
        # We need to recalculate county hotspots per year first.

        # Re-calculate county hotspots by year
        if eaglei_events is None or eaglei_events.empty or 'year' not in eaglei_events.columns or 'county_fips' not in eaglei_events.columns or 'customers_affected' not in eaglei_events.columns:
             print("Error: eaglei_events data is not loaded or missing required columns for annual county hotspots.")
             outage_customer_data_latest_year = pd.DataFrame()
        else:
            county_hotspots_annual = eaglei_events.groupby(['year', 'county_fips']).agg(
                events=('duration_min', 'count'),
                customers_affected=('customers_affected', 'sum'),
                avg_duration_min=('duration_min', 'mean')
            ).reset_index()

            # Merge annual county hotspots with EIA customer data
            outage_customer_data_annual = pd.merge(county_hotspots_annual, eia_county_year, on=['year', 'county_fips'], how='left')

            # Calculate outages per 1000 customers for annual data
            outage_customer_data_annual['outages_per_1000_customers'] = (outage_customer_data_annual['events'] / outage_customer_data_annual['total_customers']) * 1000

            # Get the latest year from the annual data
            latest_year = outage_customer_data_annual['year'].max()

            # Filter for the latest year
            outage_customer_data_latest_year = outage_customer_data_annual[outage_customer_data_annual['year'] == latest_year].copy()


    print(f"Outage data prepared for latest year ({latest_year}).")
    display(outage_customer_data_latest_year.head())

except FileNotFoundError:
    print("Error: Required data files (county_hotspots.csv or EIA data) not found.")
except KeyError as e:
    print(f"Error: Missing expected column - {e}")
except Exception as e:
    print(f"An unexpected error occurred during data preparation: {e}")
```

```python
# Copying the psps_alignment function code to add print statements
def psps_alignment(event_df, psps_df):
    # Align PSPS dates with EAGLE-I spikes (monthly)
    print("Inside psps_alignment function...")
    if event_df is None or event_df.empty or psps_df is None or psps_df.empty:
        print("Input dataframes are None or empty.")
        return None

    event_ts_col = None
    for c in ["run_start_time", "start_time", "datetime_event_began"]: # Added datetime_event_began
        if c in event_df.columns:
            event_ts_col = c
            break
    if event_ts_col is None:
        print("No valid timestamp column found in eaglei_events.")
        return None
    print(f"Using {event_ts_col} as timestamp for eaglei_events.")


    psps_ts_col = None
    for c in ["date", "datetime", "start", "start_time", "outage_start"]: # Added outage_start
        if c in psps_df.columns:
            psps_ts_col = c
            break
    if psps_ts_col is None:
        print("No valid timestamp column found in psps.")
        return None
    print(f"Using {psps_ts_col} as timestamp for psps.")

    e_df = event_df.copy()
    # Ensure the timestamp column is datetime before creating period
    e_df[event_ts_col] = pd.to_datetime(e_df[event_ts_col], errors='coerce')
    e_df = e_df.dropna(subset=[event_ts_col]) # Drop rows where timestamp is NaT
    e_df["ym"] = e_df[event_ts_col].dt.to_period("M")
    print("e_df head and shape after ym creation:")
    display(e_df.head())
    print(e_df.shape)


    p_df = psps_df.copy()
    # Ensure the timestamp column is datetime before creating period
    p_df[psps_ts_col] = pd.to_datetime(p_df[psps_ts_col], errors='coerce')
    p_df = p_df.dropna(subset=[psps_ts_col]) # Drop rows where timestamp is NaT
    p_df["ym"] = p_df[psps_ts_col].dt.to_period("M")
    print("p_df head and shape after ym creation:")
    display(p_df.head())
    print(p_df.shape)

    # Aggregate EAGLE-I customers by month
    # Ensure 'customers_affected' is numeric
    e_df['customers_affected'] = pd.to_numeric(e_df['customers_affected'], errors='coerce')
    m1 = (e_df.groupby("ym")["customers_affected"].sum(min_count=1)
          .reset_index(name="eaglei_customers"))
    print("m1 (eaglei_customers by ym):")
    display(m1.head())
    print(m1.shape)


    # Aggregate PSPS events by month
    # Assuming 'psps_event_count' is available from previous step or count rows
    if 'psps_event_count' in p_df.columns:
         m2 = p_df.groupby("ym")["psps_event_count"].sum(min_count=1).reset_index(name="psps_events")
         print("Aggregating PSPS using 'psps_event_count'.")
    else:
        m2 = p_df.groupby("ym").size().reset_index(name="psps_events")
        print("Aggregating PSPS by counting rows.")

    print("m2 (psps_events by ym):")
    display(m2.head())
    print(m2.shape)


    # Merge the two monthly aggregations
    out = m1.merge(m2, on="ym", how="outer").sort_values("ym")
    print("out (merged df) head and shape:")
    display(out.head())
    print(out.shape)

    # Save the output and return the path
    output_path = OUTDIR / "psps_alignment_monthly.csv"
    if not out.empty:
        out.to_csv(output_path, index=False)
        print(f"[saved] psps_alignment_monthly.csv  ->  {output_path}")
        return output_path
    else:
        print("Merged dataframe is empty, skipping save.")
        return None

# Filter PSPS data to align with EAGLE-I years (2014-2023) and add a count column
valid_years = eaglei_events['year'].unique()
psps_filtered = psps[psps['outage_year'].isin(valid_years)].copy()
psps_filtered['psps_event_count'] = 1 # Add a column to count PSPS events per row

# Re-run the psps_alignment function with filtered PSPS data
psps_alignment_output_path = psps_alignment(eaglei_events, psps_filtered)

# Now attempt to load and plot the data again
try:
    # Check if the function returned a valid path
    if psps_alignment_output_path is None or not Path(psps_alignment_output_path).exists():
         print("Failed to generate psps_alignment_monthly.csv.")
         psps_alignment_df = pd.DataFrame() # Create empty DataFrame to avoid errors
    else:
        psps_alignment_df = pd.read_csv(psps_alignment_output_path)

    if psps_alignment_df.empty:
        print("No valid data points for PSPS alignment after generation.")
    else:
        # Ensure 'ym' is treated as a period or datetime for plotting
        psps_alignment_df['ym'] = pd.to_datetime(psps_alignment_df['ym'])

        # Create a figure and a set of subplots with a shared x-axis
        fig, ax1 = plt.subplots(figsize=(12, 6))

        # Plot 'eaglei_customers' on the first y-axis
        ax1.plot(psps_alignment_df['ym'], psps_alignment_df['eaglei_customers'], 'b-', label='EAGLE-I Customers Affected')
        ax1.set_xlabel('Month')
        ax1.set_ylabel('EAGLE-I Customers Affected', color='b')
        ax1.tick_params(axis='y', labelcolor='b')

        # Create a second y-axis
        ax2 = ax1.twinx()

        # Plot 'psps_events' on the second y-axis
        ax2.plot(psps_alignment_df['ym'], psps_alignment_df['psps_events'], 'r-', label='Number of PSPS Events')
        ax2.set_ylabel('Number of PSPS Events', color='r')
        ax2.tick_params(axis='y', labelcolor='r')

        # Add title and legend
        plt.title('Monthly Alignment of EAGLE-I Customers Affected and PSPS Events (2014-2023)')
        fig.legend(loc="upper left", bbox_to_anchor=(0.1, 0.9))

        # Format x-axis for better readability
        plt.xticks(rotation=45, ha='right')

        plt.tight_layout()
        plt.show()

except FileNotFoundError:
    print("Error: psps_alignment_monthly.csv not found after generation.")
except KeyError as e:
    print(f"Error: Missing expected column in psps_alignment_monthly.csv - {e}")
except Exception as e:
    print(f"An unexpected error occurred while plotting PSPS alignment: {e}")
```
